### Data Preprocessing

```python



import re
import nltk
    #nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stopwords_processing():
    all_stopwords = stopwords.words('english')
    all_stopwords.remove('not')
    return all_stopwords


   

```

```python

def data_cleaning():
  corpus=[]
  st_words=stopwords_processing()
  ps = PorterStemmer()
  import re
  for i in dataset['Review']:
    review = re.sub('[^a-zA-Z]', ' ', i)
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in set(st_words)]
    review = ' '.join(review)
    corpus.append(review)
  return corpus
    
  
```

```python
corpus=data_cleaning()
```

### Data transformation

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1420)
```

```python
corpus_word=data_cleaning()
X = cv.fit_transform(corpus_word).toarray()
y = dataset.iloc[:, -1].values
```

```python
# Saving BoW dictionary to later use in prediction
import pickle
bow_path = 'c1_BoW_Sentiment_Model.pkl'
pickle.dump(cv, open(bow_path, "wb"))
```

### Dividing dataset into training and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

### Model fitting (Naive Bayes)

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
classifier2=MultinomialNB()
classifier3=DecisionTreeClassifier()
classifier4=KNeighborsClassifier()
classifier.fit(X_train, y_train)
```

```python
# Exporting NB Classifier to later use in prediction
import joblib
joblib.dump(classifier, 'Classifier_Sentiment_Model') 
```

###Model performance

```python
y_pred = classifier.predict(X_test)

from sklearn.metrics import  accuracy_score


accuracy_score(y_test, y_pred)
```
